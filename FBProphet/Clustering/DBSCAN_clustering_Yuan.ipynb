```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler


# #############################################################################
# Generate sample data
#centers = [[1, 1], [-1, -1], [1, -1]]
#X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4,
                           # random_state=0)

#X = StandardScaler().fit_transform(X)
path='D://4th Semester//Data Mining2//DMC//Dataset//'
df = pd.read_csv(path+'data4clustering.csv')
X = pd.read_csv(path+'data4clustering.csv')
X.drop(['key','size','pid','brand','color','releaseDate','rrp','size_normalization','total_sales','total_sales_range2'], axis=1, inplace=True)
X.head()
```

```python

# #############################################################################
# Compute DBSCAN
X=X.fillna(0)
db = DBSCAN(eps=1, min_samples=5).fit(X)
#core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
#core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
print(set(labels))
# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
df.insert(3,'Cluster',labels)
df.to_csv(path+'clusters_v2.csv')
df.head()
```
