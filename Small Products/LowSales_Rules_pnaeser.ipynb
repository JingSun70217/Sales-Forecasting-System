```python
#read test data
test0=pd.read_csv(test_data_directory + 'test_0.csv')
test1=pd.read_csv(test_data_directory + 'test_1.csv')
test2=pd.read_csv(test_data_directory + 'test_2.csv')
test3=pd.read_csv(test_data_directory + 'test_3.csv')
test4=pd.read_csv(test_data_directory + 'test_4.csv')
```

```python
#change type from float to int as preparation for key generation
test0['pid']=test0['pid'].astype(int)
test1['pid']=test1['pid'].astype(int)
test2['pid']=test2['pid'].astype(int)
test3['pid']=test3['pid'].astype(int)
test4['pid']=test4['pid'].astype(int)
```

```python
#create key function
def keycr(train):
    keys3=[]
    i=0
    while i<len(train):
        key=str(train.iloc[i,1])+str(train.iloc[i,2])
        keys3.append(key)
        i=i+1
    return keys3
```

```python
#creating the keys for test data
test0['key']=keycr(test0)
test1['key']=keycr(test1)
test2['key']=keycr(test2)
test3['key']=keycr(test3)
test4['key']=keycr(test4)
```

```python
df_train = df_train.reset_index(drop=True)
```

```python
# function creates new test data set which includes only the small product
def seperate_test(test):
    i=0 
    arr=[]
    c=0
    test = test.reset_index(drop=True)
    while i<len(df_train):
        if df_train['key'][i] in test['key'].values:
            arr.append(df_train['key'][i])
            c=c+1
        i=i+1
    print('NUMBER OF KEYS OF SMALL PRODUCTS IN TEST')
    print(c)
    i=0 
    test00=test
    while i<len(test):
        if test['key'][i] not in arr:
            test00=test00[test00.key != test['key'][i]]
        i=i+1
    return test00
```

```python
#applying the function
test0=seperate_test(test0)
test1=seperate_test(test1)
test2=seperate_test(test2)
test3=seperate_test(test3)
test4=seperate_test(test4)
```

```python
#objective functions from chung
def abs_sales_diff(pred, target):
    assert len(pred) == len(target)
    return sum([abs(pred[i] - target[i]) for i in range(len(pred))])
```

```python
def brute_force_mean(mean):
    factor_start = 1.0
    mean_start = mean
    best_factor = factor_start
    best_mean = mean
    best_error = 700 #just some init
    for i in range(100):
        factor = factor_start+i*0.01
        mean = mean_start*factor
        test0['global_avg']=test0['stock']/mean
        test1['global_avg']=test1['stock']/mean
        test2['global_avg']=test2['stock']/mean
        test3['global_avg']=test3['stock']/mean
        test4['global_avg']=test4['stock']/mean
        temp = np.array(test0['global_avg'].values.tolist())
        test0['global_avg'] = np.where(temp > 31, 18, temp).tolist()
        temp = np.array(test1['global_avg'].values.tolist())
        test1['global_avg'] = np.where(temp > 31, 18, temp).tolist()
        temp = np.array(test2['global_avg'].values.tolist())
        test2['global_avg'] = np.where(temp > 31, 18, temp).tolist()
        temp = np.array(test3['global_avg'].values.tolist())
        test3['global_avg'] = np.where(temp > 31, 18, temp).tolist()
        temp = np.array(test4['global_avg'].values.tolist())
        test4['global_avg'] = np.where(temp > 31, 18, temp).tolist()
        pred = test0.global_avg.astype(int)
        target=[]
        for i in range(len(test0)):
            target.append(int(test0['sold_out_date'][i][8:10]))
        result0 = np.sqrt(abs_sales_diff(pred, target))
        pred = test1.global_avg.astype(int)
        target=[]
        for i in range(len(test1)):
            target.append(int(test1['sold_out_date'][i][8:10]))
        result1 = np.sqrt(abs_sales_diff(pred, target))
        pred = test2.global_avg.astype(int)
        target=[]
        for i in range(len(test2)):
            target.append(int(test2['sold_out_date'][i][8:10]))
        result2 = np.sqrt(abs_sales_diff(pred, target))
        pred = test3.global_avg.astype(int)
        target=[]
        for i in range(len(test3)):
            target.append(int(test3['sold_out_date'][i][8:10]))
        result3 = np.sqrt(abs_sales_diff(pred, target))
        pred = test4.global_avg.astype(int)
        target=[]
        for i in range(len(test4)):
            target.append(int(test4['sold_out_date'][i][8:10]))
        result4 = np.sqrt(abs_sales_diff(pred, target))
        this_run = np.mean([result0,result1,result2,result3,result4])
        if (this_run<best_error):
            best_mean = mean
            best_error = this_run
            best_factor = factor
    #print('Mean:')
    #print (best_error)
    return best_factor, best_mean, best_error
```

# Pattern Dec, Nov, Oct

```python
#function to apply pattern to training set
def apply_pattern(test, pattern):
    stocks = test['stock']
    assert len(stocks) == len(pattern)
    pred=[]
    for i in range(len(stocks)):
        day = 0
        stock = stocks[i]
        for j in range(len(pattern[i])):
            if (pattern[i][j] > 0 and stock >0):
                stock = stock-pattern[i][j]
                day = j
        if (day == 0):
            #fallback
            day = 18
        pred.append(day)
    return pred
```

```python
#Get the sales pattern of dec
list_pattern = []
for key in test0['key']:
    row = df_train[df_train['key']==key]
    pattern = []
    for i in range(31):
        pattern.append(row.iloc[0][62+i].astype(int))
    list_pattern.append(pattern)
print(list_pattern)
```
